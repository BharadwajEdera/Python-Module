```python
import mysql.connector as connection
import pandas as pandas
import csv
```

> Establish Connection

```python
conn = connection.connect(host = "localhost" , database = "test456" , user = "root" , passwd = "Bharadwaja01" , use_pure = True )
```

```python
cur = conn.cursor()
```

```python
cur.execute("create table test456.glass(index int ,RI float , Na float , Mg float , Al float , Si float ,K float , Ca float ,  Ba float , Fe float , Class float)")
```

> in the above query

> index is a reserved word

> class is a reserved word

> make Sure that your Query doesn't have any reserved words

```python
cur.execute("create table test456.glass(index_number int(20) ,RI float(10,5) , Na float(10,5) , Mg float(10,5) , Al float(10,5) , Si float(10,5) ,K float(10,5) , Ca float(10,5) ,  Ba float(10,5) , Fe float(10,5) , Class int(5))")
```

```python
import os
os.getcwd()+"/" + "MY SQL/glass.data"
```

```python
with open(os.getcwd()+"/" + "MY SQL/glass.data" , "r") as data:
    data_csv = csv.reader(data , delimiter = "\n")
    print(data_csv)
    for i in data_csv:
        print(i)
```

> enumerate

```python
with open(os.getcwd()+"/" + "MY SQL/glass.data" , "r") as data:
    data_csv = csv.reader(data , delimiter = "\n")
    print(data_csv)
    for i in enumerate(data_csv):
        print(i)
```

```python
with open(os.getcwd()+"/" + "MY SQL/glass.data" , "r") as data:
    next(data)
    data_csv = csv.reader(data , delimiter = "\n")
    print(data_csv)
    for i in enumerate(data_csv):
        for j in i[1]:
            cur.execute("insert into test456.glass values({data})".format(data= (j)))
    print("all the data inserted")    
conn.commit()
```

> WIthout enumerate and forloop

```python
import csv
with open(os.getcwd()+"/" + "MY SQL/glass.data" , "r") as data:
    next(data)
    data_csv = csv.reader(data , delimiter = "\n")
    print(data_csv)
    for i in data_csv:
        cur.execute("insert into test456.glass values({data})".format(data= (str(i[0]))))
    print("all the data inserted")    
conn.commit()
```
