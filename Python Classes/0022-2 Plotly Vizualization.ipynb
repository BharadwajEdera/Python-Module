## Checking for Outliers using Box Plot

```python
df1 = pd.DataFrame([1,2,3,4,5,6,7,8,95,6,4,1,3,2,110,112])
df1.iplot(kind = 'box')
```

```python
import scipy.stats as stats
df1 = pd.DataFrame([1,2,3,4,5,6,7,8,95,6,4,1,3,2,110,112])
# df1.iplot(kind = 'box')
stats.iqr(df1)
```
