> collection :: table

> record :: Data

> insert record into collection :: insert data into table

```python
from flask import Flask, render_template, request, jsonify
import pymongo
import pandas as pd
import csv
import json
import sys

app = Flask(__name__)

### Client Creation



@app.route('/mongoDB/database_creation', methods=['POST'])  # for calling the API from Postman/SOAPUI
def database_creation():
    if (request.method == 'POST'):
        client_url = request.json['client_url']
        client_name = request.json['client_name']

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        a = "Client Created"

        return jsonify(a)


### Collection Creation



@app.route('/mongoDB/collection_creation', methods=['POST'])  # for calling the API from Postman/SOAPUI
def collection_creation():
    if (request.method == 'POST'):
        client_url = request.json['client_url']
        client_name = request.json['client_name']
        collection_name = request.json['collection_name']

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        dataBase[collection_name]
        a = "Collection Created"

        return jsonify(a)


### Insert a Record



@app.route('/mongoDB/insert_record', methods=['POST'])  # for calling the API from Postman/SOAPUI
def insert_record():
    if (request.method == 'POST'):
        client_url = request.json['client_url']
        client_name = request.json['client_name']
        collection_name = request.json['collection_name']
        record = request.json['record']

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        collection = dataBase[collection_name]
        collection.insert_one(record)
        a = "Record Inserted"
        b = client.list_database_names()

        return jsonify(a, b)


### insert CSV file into mongodb



@app.route('/mongoDB/insert_csv', methods=['POST'])  # for calling the API from Postman/SOAPUI
def insert_csv():
    if (request.method == 'POST'):
        client_url = request.json['client_url']
        client_name = request.json['client_name']
        collection_name = request.json['collection_name']

        #path = request.json['path']
        file_name = request.json['file_name']

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        collection = dataBase[collection_name]

        df = pd.read_csv( file_name + '.csv',encoding = 'ISO-8859-1')   # loading csv file
        df.to_json( file_name + '.json')                               # saving to json file
        jdf = open( file_name + '.json').read()                        # loading the json file 
        data = json.loads(jdf)  

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        collection = dataBase[collection_name]
        collection.insert_one(data)

        a = "Record Inserted"
        b = client.list_database_names()

        return jsonify(a, b)
    
### Download Document from mongodb and save it in  CSV file 
@app.route('/mongoDB/download', methods=['POST'])  # for calling the API from Postman/SOAPUI
def download():
    if (request.method == 'POST'):
        client_url = request.json['client_url']
        client_name = request.json['client_name']
        collection_name = request.json['collection_name']

        client = pymongo.MongoClient(client_url)
        dataBase = client[client_name]
        collection = dataBase[collection_name]
        data = collection.find_one()
        return jsonify(data)


if __name__ == '__main__':
    app.run()
```

## Create Keyspace

```python
from flask import Flask, render_template, request, jsonify

from cassandra.cluster import Cluster
import csv
cluster = Cluster()
session_offline = cluster.connect()

app = Flask(__name__)

#create keyspace:

@app.route('/create_keyspace', methods=['POST']) # for calling the API from Postman/SOAPUI
def create_keyspace():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        strategy_name = request.json['strategy_name']
        replication_factor = request.json['replication_factor']
        row = session_offline.execute("CREATE KEYSPACE " + str(keyspace_name) + " WITH replication = {'class':" + "'" + str(strategy_name) + "'" + "," + "'replication_factor':" + "'" + str(replication_factor) + "'}").one()
        a = "keyspace created"
        return jsonify(row, a)
    

#create table:

@app.route('/create_table', methods=['POST']) # for calling the API from Postman/SOAPUI
def create_table():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        table_name = request.json['table_name']
        column1 = request.json['column1']
        dtype1 = request.json['dtype1']
        constraint_name = request.json['constraint_name']
        column2 = request.json['column2']
        dtype2 = request.json['dtype2'] 
        row = session_offline.execute("CREATE TABLE " + str(keyspace_name) + "." + str(table_name) + "(" + str(column1) + " " + str(dtype1) + " " + str(constraint_name) + "," + str(column2) + " " + str(dtype2) + ");").one()
        a = "table created"
        return jsonify(a)


    
    
#insert data into table:

@app.route('/insert_into_table', methods=['POST']) # for calling the API from Postman/SOAPUI
def insert_into_table():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        table_name = request.json['table_name']
        column1 = request.json['column1']
        column2 = request.json['column2']
        value1 = request.json['value1']
        value2 = request.json['value2']
        row = session_offline.execute("INSERT INTO " + str(keyspace_name) + "." + str(table_name) + "(" + str(column1) + "," + str(column2) + ") " "values(" + str(value1) + "," + str(value2) + ");").one()
        a = "data inserted"
        return jsonify(a)
    
    
    
# Delete record:

@app.route('/delete_from_table', methods=['POST']) # for calling the API from Postman/SOAPUI
def delete_from_table():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        table_name = request.json['table_name']
        column_name = request.json['column_name']
        column_value = request.json['column_value']
        row = session_offline.execute("DELETE from " + str(keyspace_name) + "." + str(table_name) + " " + "where " + str(column_name) + "=" + str(column_value) + ";").one()
        a = "record deleted"
        return jsonify(a)
    
    
# Update table:

@app.route('/update_record', methods=['POST']) # for calling the API from Postman/SOAPUI
def update_record():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        table_name = request.json['table_name']
        column_name = request.json['column_name']
        value = request.json['value']
        column_name2 = request.json['column_name2']
        value2 = request.json['value2']
        row = session_offline.execute("UPDATE " + str(keyspace_name) + "." + str(table_name) + " " + "set " + str(column_name) + "=" + str(value) + " where " + str(column_name2) + "=" + str(value2) + ";").one()
        a = "record updated"
        return jsonify(a)
    
    
# Bulk inserstion into table:

@app.route('/bulk_insertion', methods=['POST']) # for calling the API from Postman/SOAPUI
def bulk_insertion():
    if (request.method=='POST'):
        keyspace_name = request.json['keyspace_name']
        table_name = request.json['table_name']
        column1 = request.json['column1']
        dtype = request.json['dtype']
        constraint_name = request.json['constraint_name']
        column2 = request.json['column2']
        column3 = request.json['column3']
        column4 = request.json['column4']
        file_name = request.json['file_name']
        
        
        
        row = session_offline.execute("CREATE TABLE " + str(keyspace_name) + "." + str(table_name) + "(" + str(column1) + " " + str(dtype) + " " + str(constraint_name) + "," + str(column2) + " " + str(dtype) + "," + str(column3) + " " + str(dtype) + "," + str(column4) + " " + str(dtype) + ")" + ";").one()
        a = "table created"
        
        

        csv_data = csv.reader(open( file_name + '.csv'))
        header = next(csv_data)
        for i in csv_data:
            row = session_offline.execute("INSERT INTO " + str(keyspace_name) + "." + str(table_name) + "(" + str(column1) + "," + str(column2) + "," + str(column3) + "," + str(column4) + ") VALUES (%s , %s , %s , %s);" , i).one()
        b = "Data from CSV is inserted into Table"
        
        
        return jsonify(a,b)
    
    
    

if __name__ == '__main__':
    app.run()
```
