### Creating a Keyspace in the Cassandra Cloud

### Keyspace is Similar to Table

### Creating a Keyspace in the Cassandra Local
https://docs.datastax.com/en/cql-oss/3.1/cql/cql_reference/create_keyspace_r.html
### Copy the below command at the cqlsh
### CREATE KEYSPACE KeySpaceName  WITH replication = {'class':'SimpleStrategy','replication_factor':'1'} AND durable_writes='true';

### Example :
### CREATE KEYSPACE Bharadwaja01  WITH replication = {'class':'SimpleStrategy','replication_factor':'1'} AND durable_writes='true';
