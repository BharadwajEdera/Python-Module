### Question 1. 
>For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme:




>Examples

>meme_sum(26, 39) ➞ 515

>\# 2+3 = 5, 6+9 = 15

>\# 26 + 39 = 515

>meme_sum(122, 81) ➞ 1103

>\# 1+0 = 1, 2+8 = 10, 2+1 = 3

>\# 122 + 81 = 1103

>meme_sum(1222, 30277) ➞ 31499



```python
def meme_sum(n1 , n2):
    l = []
    s1 = str(n1)
    s2 = str(n2)
    if len(s1) <= len(s2):
        k1 = "0"*(len(s2) - len(s1)) + s1
        k2 = str(s2)
    elif len(s1) >= len(s2):
        k1 = str(s1)
        k2 = "0"*(len(s1) - len(s2)) + s2
    else:
        k1 = str(s1)
        k2 = str(s2)
    for i in range(len(k1)):
        l.append(str(int(k1[i])+int(k2[i])))
    
    return int("".join(l))
```

```python
meme_sum(26, 39)
```

```python
meme_sum(122, 81)
```

```python
meme_sum(1222, 30277)
```

### Question 2. 
>Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.

>Examples

>next_prime(12) ➞ 13

>next_prime(24) ➞ 29

>next_prime(11) ➞ 11

>\# 11 is a prime, so we return the number itself.



```python
def next_prime(n):
    while True:
        for i in range(2,int(n/2)+1):
            if n%i == 0:
                n = n+1
                break
        else:
            return n
```

```python
next_prime(12)
```

```python
next_prime(24)
```

```python
next_prime(11) 
```

### Question 4. 
>The Kempner Function, applied to a composite number, permits to find the smallest integer greater than zero whose factorial is exactly divided by the number.

>kempner(6) ➞ 3

>1! = 1 % 6 > 0

>2! = 2 % 6 > 0

>3! = 6 % 6 === 0

>kempner(10) ➞ 5

>1! = 1 % 10 > 0

>2! = 2 % 10 > 0

>3! = 6 % 10 > 0

>4! = 24 % 10 > 0

>5! = 120 % 10 === 0

>A Kempner Function applied to a prime will always return the prime itself.

>kempner(2) ➞ 2

>kempner(5) ➞ 5

>Given an integer n, implement a Kempner Function.

>Examples

>kempner(6) ➞ 3

>kempner(10) ➞ 5

>kempner(2) ➞ 2



```python
def fact(n):
    if n in (0,1):
        return 1
    else:
        return n*fact(n-1)
    
def kempner(n):
    for i in range(n):
        if fact(i)%n == 0:
            return i
        else:
            continue
    else:
        return n
    
```

```python
kempner(10)
```

```python
kempner(2)
```

```python
kempner(5)
```

### Question 5. 
>You work in a factory, and your job is to take items from a conveyor belt and pack them into boxes. Each box can hold a maximum of 10 kgs. Given a list containing the weight (in kg) of each item, how many boxes would you need to pack all of the items?

>Example

>boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2]) ➞ 5

>\# Box 1 = [2, 1, 2, 5] (10kg)

>\# Box 2 = [4, 3] (7kg)

>\# Box 3 = [6, 1, 1] (8kg)

>\# Box 4 = [9] (9kg)

>\# Box 5 = [3, 2] (5kg)


```python
def boxes(l):
    import numpy as np
    import math
    n = np.array(l)
    print(np.sum(n))
    print((np.sum(n)/10))
    return math.ceil((np.sum(n)/10))
```
